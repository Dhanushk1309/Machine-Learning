```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("house_price.csv")
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
#y=mx+b
135.78767123*3300+180616.43835616432
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area(sq.ft)')
plt.ylabel('price(rs)')
plt.scatter(df.area,df.price,color='blue' , marker='*')
plt.plot(df.area,reg.predict(df[['area']]), color='red')
plt.show()
```

```python
d=pd.read_csv('areas.csv')
d.head(3)
```

```python
p=reg.predict(d.values)
```

```python
d['prices']=p
```

```python
d.to_csv("predictions.csv",index=False)
```

```python
pf=pd.read_csv("canada_per_capita_income.csv")
pf.head(2)
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.scatter(pf.year,pf['per capita income (US$)'])
plt.show()
```

```python
reg1=linear_model.LinearRegression()
reg1.fit(pf[['year']],pf['per capita income (US$)'])
```

```python
reg1.predict([[2020]])
```

```python
reg1.coef_
```

```python
reg1.intercept_
```

```python
#price
135.78767123*2020+180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.scatter(pf.year,pf['per capita income (US$)'],color='red')
plt.plot(pf.year,reg1.predict(pf[['year']]), color='blue')
plt.show()
```
